```python
%%time
import re

text = ' '.join(df['twitts'])
text = text.split()
freq_comm = pd.Series(text).value_counts()
rare = freq_comm[freq_comm.value == 1]

def get_clean_text(x):
    if type(x) is str:
        x = x.lower()
        for key in contractions:
            value = contractions[key]
            x = x.replace(key, value)
    
    return x
text = df['twitts'].tolist()
text[:3]
```

```python
y = df['sentiment']
token= Tokenizer()
token.fit_on_texts(text)
```

```python
vocab_size = len(token.word_index)+ 1
vocab_size
```

```python
print(token.index_word)
```

```python
encoded_text = token.texts_to_sequences(text)
print(encoded_text[:3])
```

```python
max_len = 120
X = pad_sequences(encoded_text, maxlen = max_len, padding='post')
print(X)
```

```python
X.shape
```

```python
glove_vectors = dict()


```

```python
%%time
file = open('C://Users/rschm/OneDrive/Documents/glove.twitter.27B.100d.txt', 'r', encoding = 'utf-8')
trial = []
for line in file:
    values = line.split()
    word = values[0]
    vectors = np.asarray(values[1:])
    glove_vectors[word] = vectors
    trail.append(values[1:])
    
file.close()

```

```python
len(glove_vectors.keys())
```

```python
glove_vectors.get('he').shape
```

```python
word_vector_matrix = np.zeros((vocab_size, 100))

for word, index in token.word_index.items():
    vector = glove_vectors.get(word)
    if vector is not None:
        word_vector_matrix[index] = vector
    else:
        print(word)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 42, 
                                                    test_size = 0.2, stratify = y)
```

```python
vec_size = 100

model = Sequential()
model.add(Embedding(
vocab_size, 
vec_size, 
input_length = max_len,
weights = [word_vector_matrix],
trainable = False
))
model.add(Conv1D(64, 8, activation = 'relu'))
model.add(MaxPooling1D(2))
model.add(Dropout(0.5))

model.add(Dense(32, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(16, activation='relu'))

model.add(GlobalMaxPooling1D())

model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer=Adam(learning_rate = 0.0001), 
              loss = 'binary_crossentropy',
              metrics = ['accuracy']
             )

model.fit(X_train, y_train, epochs=30, validation_data = (X_test, y_test) )
```
